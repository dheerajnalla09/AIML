Sentiment Analysis

**What is Sentiment Analysis?**  
It's the process of determining whether text expresses positive, negative, or neutral emotions.

**Use Cases:**
- Customer review analysis
- Social media monitoring
- Brand reputation tracking
- Product feedback analysis

```python
# Import the pipeline function from transformers
from transformers import pipeline

# Create a sentiment analysis pipeline
# This automatically downloads and loads a pre-trained model
print("Loading sentiment analysis model...")
sentiment_analyzer = pipeline("sentiment-analysis")

print("✅ Model loaded successfully!")
print(f"Model: {sentiment_analyzer.model.name_or_path}")
```

### Test the Sentiment Analyzer

Let's test it with various sentences to see how it works.

```python
# Test with a positive sentence
text1 = "I absolutely love this product! It's amazing!"
result1 = sentiment_analyzer(text1)

print(f"Text: {text1}")
print(f"Result: {result1}")
print(f"Sentiment: {result1[0]['label']}")
print(f"Confidence: {result1[0]['score']:.2%}")
print("-" * 60)
```

**Explanation:**  
- The model correctly identifies this as POSITIVE sentiment
- The score (0.9998) shows 99.98% confidence
- Higher score = more confident the model is


```python
# Test with multiple sentences at once
test_sentences = [
    "This is the worst experience ever!",
    "The product is okay, nothing special.",
    "Best purchase I've made this year!",
    "I'm disappointed with the quality.",
    "Exactly what I needed, perfect!"
]

print("📊 Analyzing multiple sentences:\n")
for sentence in test_sentences:
    result = sentiment_analyzer(sentence)[0]
    emoji = "😊" if result['label'] == 'POSITIVE' else "😞"
    print(f"{emoji} {sentence}")
    print(f"   → {result['label']} ({result['score']:.2%})")
    print()
```

```python
for sentence in test_sentences:
    result = sentiment_analyzer(sentence)[0]

    if result['score'] < 0.75:
        label = "NEUTRAL"
        emoji = "😐"
    else:
        label = result['label']
        emoji = "😊" if label == "POSITIVE" else "😞"

    print(f"{emoji} {sentence}")
    print(f"   → {label} ({result['score']:.2%})")
    print()
```

 Text Generation

**What is Text Generation?**  
AI creates new text by predicting what words should come next based on the input prompt.

**Use Cases:**
- Story writing
- Code completion
- Email drafting
- Content creation

```python
# Load a text generation model
print("Loading text generation model (this may take a minute)...")
text_generator = pipeline("text-generation", model="distilgpt2")

print("✅ Text generator loaded!")
print(f"Model: {text_generator.model.name_or_path}")
```

```python
# Generate text from a simple prompt
prompt = "Artificial intelligence is"

result = text_generator(
    prompt,
    max_length=50,        # Maximum number of tokens (words) to generate
    num_return_sequences=1,  # How many different versions
    temperature=0.7       # Creativity (0.0 = focused, 1.0 = creative)
)

print(f"Prompt: {prompt}")
print(f"\nGenerated Text:")
print(result[0]['generated_text'])
```

```python
# Let's see the effect of different parameters
prompt = "The future of technology will"

print("🎯 Testing Different Settings:\n")
print("=" * 70)

# Setting 1: Short, focused response
print("\n1️⃣ SHORT & FOCUSED (max_length=30, temperature=0.3)")
result1 = text_generator(prompt, max_length=30, temperature=0.3, num_return_sequences=1)
print(result1[0]['generated_text'])

# Setting 2: Medium, balanced response
print("\n2️⃣ MEDIUM & BALANCED (max_length=50, temperature=0.7)")
result2 = text_generator(prompt, max_length=50, temperature=0.7, num_return_sequences=1)
print(result2[0]['generated_text'])

# Setting 3: Long, creative response
print("\n3️⃣ LONG & CREATIVE (max_length=80, temperature=0.9)")
result3 = text_generator(prompt, max_length=80, temperature=0.9, num_return_sequences=1)
print(result3[0]['generated_text'])
```

```python
# Generate multiple variations of the same prompt
prompt = "Machine learning helps us"

print(f"Prompt: {prompt}\n")
print("Generating 3 different completions...\n")

results = text_generator(
    prompt,
    max_length=40,
    num_return_sequences=3,  # Generate 3 different versions
    temperature=0.8
)

for idx, result in enumerate(results, 1):
    print(f"Version {idx}:")
    print(result['generated_text'])
    print("-" * 70)
```

```python
from transformers import pipeline

translator_en_to_fr = pipeline(
    "text2text-generation",
    model="Helsinki-NLP/opus-mt-en-fr"
)

sentences_to_translate = [
    "Hello, how are you?",
    "I love learning about artificial intelligence.",
    "The weather is beautiful today.",
    "Where is the nearest restaurant?"
]

print("🇬🇧 → 🇫🇷 English to French Translation:\n")

for sentence in sentences_to_translate:
    # 🔥 IMPORTANT: Add prefix
    text = f">>fr<< {sentence}"

    result = translator_en_to_fr(text, max_new_tokens=50)

    print(f"English:  {sentence}")
    print(f"French:   {result[0]['generated_text']}")
    print()
```

```python
from transformers import pipeline

print("Loading translators...")

# French model
translator_en_to_fr = pipeline(
    "text2text-generation",
    model="Helsinki-NLP/opus-mt-en-fr"
)

# German model
translator_en_to_de = pipeline(
    "text2text-generation",
    model="Helsinki-NLP/opus-mt-en-de"
)

text = "Machine learning is transforming the world."

print(f"\n🇬🇧 English: {text}")

# Translate to French
french = translator_en_to_fr(f">>fr<< {text}", max_new_tokens=50)[0]['generated_text']
print(f"🇫🇷 French:  {french}")

# Translate to German
german = translator_en_to_de(f">>de<< {text}", max_new_tokens=50)[0]['generated_text']
print(f"🇩🇪 German:  {german}")
```

 Build a Simple AI Assistant

Combine all three capabilities into one assistant!


```python
class SimpleAIAssistant:
    """A simple AI assistant with multiple capabilities"""

    def __init__(self):
        print("Initializing AI Assistant...")
        self.sentiment_analyzer = pipeline("sentiment-analysis")
        self.text_generator = pipeline("text-generation", model="distilgpt2")
        self.translator = pipeline("translation_en_to_fr")
        print("✅ Assistant ready!\n")

    def analyze_sentiment(self, text):
        """Analyze sentiment of text"""
        result = self.sentiment_analyzer(text)[0]
        return f"{result['label']} ({result['score']:.2%})"

    def generate_text(self, prompt, length=50):
        """Generate text from prompt"""
        result = self.text_generator(prompt, max_length=length, num_return_sequences=1)
        return result[0]['generated_text']

    def translate(self, text):
        """Translate English to French"""
        result = self.translator(text)[0]
        return result['translation_text']

# Create the assistant
assistant = SimpleAIAssistant()
```

```python
# Test the assistant
print("🤖 AI Assistant Demo\n")
print("="*70)

# Test 1: Sentiment Analysis
text1 = "I'm so excited about learning AI!"
print(f"\n📝 Text: {text1}")
print(f"😊 Sentiment: {assistant.analyze_sentiment(text1)}")

# Test 2: Text Generation
prompt = "The most important skill for the future is"
print(f"\n✍️ Prompt: {prompt}")
print(f"📄 Generated: {assistant.generate_text(prompt, length=40)}")

# Test 3: Translation
text2 = "Welcome to the world of AI!"
print(f"\n🌐 English: {text2}")
print(f"🇫🇷 French: {assistant.translate(text2)}")

print("\n" + "="*70)
print("✅ All features working!")
```
