# Flight Fare Prediction using SVR and Random Forest
This notebook demonstrates how to predict flight fares using two machine learning models:
- **Support Vector Regression (SVR)**
- **Random Forest Regressor**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## Step 1: Load Dataset

```python
data = {
    'airline': ['SpiceJet', 'SpiceJet', 'AirAsia', 'Vistara'],
    'flight': ['SG-8709', 'SG-8157', 'I5-764', 'UK-995'],
    'source_city': ['Delhi', 'Delhi', 'Delhi', 'Delhi'],
    'departure_time': ['Evening', 'Early_Morning', 'Early_Morning', 'Morning'],
    'stops': ['zero', 'zero', 'zero', 'zero'],
    'arrival_time': ['Night', 'Morning', 'Early_Morning', 'Afternoon'],
    'destination_city': ['Mumbai', 'Mumbai', 'Mumbai', 'Mumbai'],
    'class': ['Economy', 'Economy', 'Economy', 'Economy'],
    'duration': [2.17, 2.33, 2.17, 2.25],
    'days_left': [1, 1, 1, 1],
    'price': [5953, 5953, 5956, 5955]
}
df = pd.DataFrame(data)
df.head()
```

```python
categorical_cols = ['airline','flight','source_city','departure_time','stops','arrival_time','destination_city','class']
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
df.head()
```

## Step 3: Define Features and Target

```python
X = df.drop(columns=['price'])
y = df['price']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
## Step 4: Train Models
```

```python
svr = SVR(kernel='rbf', )
svr.fit(X_train, y_train)
y_pred_svr = svr.predict(X_test)

rf = RandomForestRegressor(random_state=42,n_estimators=100, max_depth=5)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

## Step 5: Evaluate Models

```python
print('SVR Results:')
print('MAE:', mean_absolute_error(y_test, y_pred_svr))
print('MSE:', mean_squared_error(y_test, y_pred_svr))
print('R2:', r2_score(y_test, y_pred_svr))

print('\nRandom Forest Results:')
print('MAE:', mean_absolute_error(y_test, y_pred_rf))
print('MSE:', mean_squared_error(y_test, y_pred_rf))
print('R2:', r2_score(y_test, y_pred_rf))
```
