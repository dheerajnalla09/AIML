5. Merge two dictionaries; if keys overlap, sum their values.

````python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}

for k in d2:
    if k in d1:
        d1[k] = d1[k] + d2[k]
    else:
        d1[k] = d2[k]

print(d1)
````

7. Given a string, count vowels, consonants, digits, and special characters.

````python
s = "Hello123@!"

v = c = d = sp = 0

for ch in s:
    if ch.isalpha():
        if ch.lower() in "aeiou":
            v += 1
        else:
            c += 1
    elif ch.isdigit():
        d += 1
    else:
        sp += 1

print("Vowels:", v)
print("Consonants:", c)
print("Digits:", d)
print("Special:", sp)
````

2. Write a recursive function to compute factorial.

````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# check
print(factorial(5))
````

6. Write a function to return all prime numbers up to n.

````python
def primes_upto(n):
    primes = []

    for num in range(2, n + 1):
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            primes.append(num)

    return primes

# Example
print(primes_upto(20))
````

2. Add a method to calculate grade in Student class.

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "Fail"

# Example
s1 = Student("John", 82)

print(s1.name)
print(s1.marks)
print("Grade:", s1.grade())
````

3. Create class Rectangle with area and perimeter methods.

````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Example
r = Rectangle(5, 3)

print("Area:", r.area())
print("Perimeter:", r.perimeter())
````

6. Create class BankAccount with deposit and withdraw methods.

````python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

# Example
acc = BankAccount(1000)

acc.deposit(500)
acc.withdraw(300)
print("Balance:", acc.balance)
````

3. Use try-except-else-finally in a program.

````python
try:
    a = int(input("Enter number: "))
    b = int(input("Enter divisor: "))

    result = a / b

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input")

else:
    print("Result:", result)

finally:
    print("Program finished")
````
